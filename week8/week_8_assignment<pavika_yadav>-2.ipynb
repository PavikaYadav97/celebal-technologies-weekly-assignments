## Tool Descriptions

### Calculator Tool
This tool evaluates mathematical expressions entered by the user and returns the computed result. It also handles invalid expressions through exception handling.

### Keyword Extraction Tool
This tool extracts important keywords from a given sentence by selecting unique words with more than four characters. It helps identify the key terms in a text.

````python
#TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
#TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower().strip(".,!?") for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## Agent Logic

The smart assistant follows a rule-based routing approach.

- If the query contains **"calculate"**, it invokes the Calculator Tool.
- If the query contains **"keywords"**, it invokes the Keyword Extraction Tool.
- Otherwise, it provides a general response.

The agent first analyzes the user query, determines the intent, invokes the appropriate tool if required, and finally returns a structured JSON response
The output is returned in a structured JSON format for consistency.

````python
def agent(query: str):
    query_lower = query.lower()
    try:
        # Calculator Route
        if "calculate" in query_lower:
            expression = query_lower.replace("calculate", "").strip()
            result = calculator(expression)
            return {
                "type": "calculation",
                "result": result
            }
        # Keyword Extraction Route
        elif "keywords" in query_lower:
            text = query.replace("Extract keywords from", "").replace("extract keywords from", "").strip()
            result = extract_keywords(text)
            return {
                "type": "keywords",
                "result": result
            }
        elif "machine learning" in query_lower:
          return {
               "type": "general",
               "result": "Machine learning is a branch of Artificial Intelligence that enables computers to learn patterns from data and make predictions without being explicitly programmed."
                }
        elif "artificial intelligence" in query_lower:
          return {
                "type": "general",
                "result": "Artificial Intelligence is the simulation of human intelligence by machines to perform tasks such as learning, reasoning, and decision making."
    }
        # General Route
        else:
            return {
                "type": "general",
                "result": f"I understand your query: '{query}'. Currently I can perform calculations and keyword extraction."
            }
    except Exception as e:
        return {
            "type": "error",
            "result": str(e)
        }
````

## Expected Output Format

The assistant returns responses in the following JSON format:

{
    "type": "calculation / keywords / general / error",
    "result": ...
}

````python
# Test Cases

queries = [
    "Calculate 20 + 5",
    "Calculate 45 * 6",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "Extract keywords from Machine learning improves decision making",
    "What is machine learning?",
    "Tell me a joke",
    "Calculate 5/0"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````
